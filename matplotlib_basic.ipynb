Matplotlib is a free Python library for creating charts and graphs like line plots, bar charts, scatter plots, and histograms. It makes data easy to visualize and understand.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Functional Approach using matplotib.pyplot

```python
#functional approach
x=np.arange(0,10)
y = x * 2
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("y = 2x plot")
plt.xlim(0,6)
plt.ylim(0,12)

#To save
plt.savefig("plot1.png")
```

Create a line plot showing the squares of numbers from 1 to 10.

```python
a = np.arange(1,11) #creating numpy array from 1 t0 10
b = a * a #array of
plt.plot(a, b)
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Line plot,b = a^2 ")
plt.xlim(1,10)
plt.ylim(1,100)
```

Plot two graphs on the same figure:

The first line should represent squares of numbers 1 to 5, in purple color.

The second line should represent a straight line (y = x), in green color.

Add a legend that clearly labels both lines.

Display the figure.

```python

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y1,color="purple",label="Squares")   # squares
plt.plot(x, y2,color="green",label="line")   # straight line
plt.legend() ##it represents label with corresponding color
plt.show()
```

##### Bar Chart
Create a bar chart showing the number of students in each subject:

Math = 40

English = 30
History = 20

```python
a = ["Math" , "Science", "English", "History"]
b = [40,35,30,20]
plt.bar(a,b,color="hotpink")
plt.xlabel("Subjects")
plt.ylabel("Score")
plt.title("Bar graph of students score")
plt.show()
```

#### Histogram

Create a histogram for exam scores:
[45, 67, 89, 45, 56, 67, 78, 88, 90, 56, 67, 76, 89, 90, 45, 56]

```python
score = [45, 67, 89, 45, 56, 67, 78, 88, 90, 56, 67, 76, 89, 90, 45, 56]
plt.hist(score,bins=10,color= "lightblue", edgecolor="black",alpha=0.5, orientation="horizontal")
plt.ylabel("Scores")
plt.xlabel("Frequency")
plt.show()
```

#### Scatter Plot
Plot a scatter plot of student study hours vs. marks:
Hours = [2, 3, 4, 5, 6, 7, 8]
Marks = [50, 55, 60, 65, 70, 75, 80]

```python
## Scatter Plot
Hours = [2,3,4,5,6,7,8]
Marks = [50,55,45,65,80,95,98]
plt.xlabel("Hours")
plt.ylabel("Score gained")
plt.title("Representing the duration of study with their marks")
plt.scatter(Hours,Marks,marker= "*") #marker could be "s"square, "o"circle(by default),(^) traingle, (+) plus, "p" pentagon, "D" Diamond
plt.show()
```

#### Pie Chart

Create a pie chart showing time spent in a day:
Sleep = 8 hrs
Study = 5 hrs
Play = 3 hrs
Others = 8 hrs

```python
#Pie Chart
activities = ["Sleep", "Study", "Play", "Others"]
duration = [8,5,3,8]
colors = ["gray", "lightblue","purple","yellow"]
explode = (0,0.5,0.1,0)
plt.pie(duration,labels= activities,colors=colors, explode= explode,autopct = "%1.1f%%")
plt.show()
```
